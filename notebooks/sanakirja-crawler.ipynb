```python
import requests
import time
import csv
from bs4 import BeautifulSoup
```

NOTE TO SELF: d -> has been manually set to utf-8. Might impact loading the files.

```python
# Initialize lists to contain all words
words = []
word_forms = []


# letters = "abcdefghijklmnopqrstuvwxyzäö"
letters = "efghijklmnopqrstuvwxyzäö"

for letter in letters:
    print(f'[INFO] Processing letter {letter}')
    
    # Define the url
    for i in range(1, 300):
        url = f'https://www.suomisanakirja.fi/selaa.php?k={letter}&p={i}'
        print(f"[INFO] Processing {url}")

        # Request and parse the HTML
        soup = BeautifulSoup(requests.get(url).content, 'html.parser')

        # Find matching div elements
        divs = soup.find_all('div', attrs={'style': 'width:300px;float:left;'})
        
        # Break out of loop if no results. We've found the last page.
        if len(divs) == 0:
            print(f'Letter {letter} processed. Last page was: {i - 1}')
            break

        # Wait n seconds before continuing to single words
        time.sleep(5)
            
        # Process each word found on the page
        for div in divs:
            # Get the text content inside div element
            word = div.string
            word_http = word.replace(' ', '%20')

            # Generate url
            url = 'https://www.suomisanakirja.fi/' + str(word_http)

            # Request and parse the HTML
            soup = BeautifulSoup(requests.get(url).content, 'html.parser')

            # Get the p element containing all 'taivutusmuodot'
            p = soup.find('p', attrs={'class': 'taiv'})

            # Wait 5 seconds between each word
            time.sleep(1)
            
            # The paragraph with class taiv does not exist. Skip the page.
            if p is None:
                continue

            # Get all the ems. These are the headings for 'taivutusmuodot' e.g. '<em>yks. nom.</em>'
            ems = p.find_all('em')

            if len(ems) > 0:
                # Convert the ems to list of strings e.g. ['yks. nom.', 'yks.gen.']
                em_list = [em.string for em in ems]
                
                # Drop None values if found
                if None in em_list:
                    em_list = [i for i in em_list if i] 

                # Convert the list into string
                em_list_str = ';'.join(em_list)
            else:
                # Use a debug value for empty lists.
                em_list_str = 'EMPTY_EM'

            # Word and the forms passed the scrutiny. Add to lists
            words.append(word)
            word_forms.append(em_list_str)
            
    # If we made it this far, let's add all the words so far into a CSV file
    with open(f'letterdata/{letter}.csv', 'w', encoding='utf-8') as f:
        print(f'[INFO] Writing into {letter}.csv')
        writer = csv.writer(f)
        writer.writerows(zip(words, word_forms))
        
    # Empty the lists. This could be done in the beginning of the loop, but
    words = []
    word_forms = []
```
